EL conjunto de datos se denomina `Automobile_data.csv`, el cual contiene información tal como: compañia, precio, kilometraje, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import os
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","Automobile_data.csv")).set_index('index')
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Elimine los valores nulos (Nan)

```python
#cambia los valores 'Nan' del Dataframe por vacios
df.replace('Nan','')
```

2. Encuentra el nombre de la compañía de automóviles más cara

```python
#Compañia más cara en cuanto al promedio del costo de sus coches
grupo=df.groupby('company')
df_aux=grupo[['price']].mean().reset_index()#promedio del costo de coche por cada marca
for i in df_aux.index:
    if df_aux.price[i]==df_aux.price.max():
        print ('En cuanto a promedio de precios, la marca mas cara es')
        print(df_aux.company[i])
```

```python
df.columns = df.columns.str.lower().str.replace('-','_').str.replace(' ','_')
df.head()
```

3. Imprimir todos los detalles de Toyota Cars

```python
#detalles de los autos Toyota
df.columns = df.columns.str.lower().str.replace('-','_').str.replace(' ','_')
df_toy=pd.DataFrame()
bs=[]
wh=[]
length=[]
et=[]
noc=[]
hp=[]
am=[]
price=[]
for i in df.index:
    if df.company[i]=='toyota':
        bs.append(df.body_style[i])
        wh.append(df.wheel_base[i])
        length.append(df.length[i])
        et.append(df.engine_type[i])
        noc.append(df.num_of_cylinders[i])
        hp.append(df.horsepower[i])
        am.append(df.average_mileage[i])
        price.append(df.price[i])
df_toy['body_style']=bs
df_toy['wheel_base']=wh
df_toy['length']=length
df_toy['engine_type']=et
df_toy['num_of_cylinders']=noc
df_toy['hp']=hp
df_toy['average_mileage']=am
df_toy['price']=price
df_toy
```

4. Cuente el total de automóviles por compañía

```python
grupo=df.groupby('company')
df_aux=grupo[['price']].count().reset_index()
for i in df_aux.index:
    print(df_aux.company[i], 'tiene', df_aux.price[i] ,'automoviles')
```

5. Encuentra el coche con el precio más alto por compañía

```python
grupo=df.groupby(['company'])
df_empty=pd.DataFrame()
df_aux=grupo[['price']].max().reset_index()
print('Estos son los coches de mas alto precio por compañia')
df_aux
```

6. Encuentre el kilometraje promedio (**average-mileage**) de cada compañía automotriz

```python
grupo=df.groupby(['company'])
grupo[['average_mileage']].mean().reset_index()
```

7. Ordenar todos los autos por columna de precio (**price**)

```python
#Vehiculos ordenados del mas barato al mas caro
df.sort_values('price')
```

```python
#Se crean DataFrame
alemanes=pd.DataFrame({'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'],
                        'Price': [23845, 171995, 135925 , 71400]})
japoneses=pd.DataFrame({'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 
                        'Price': [29995, 23600, 61500 , 58900]})
```

```python
frames=[alemanes,japoneses]
nuevo=pd.concat(frames).reset_index()
lista1=[]
lista2=[]
for j in alemanes.Company:
    lista2.append(j)
for i in nuevo.index:
        if nuevo.Company[i]in lista2:
            lista1.append('Germany')
        else:
            lista1.append('Japan')
nuevo['Country']=lista1
nuevo.drop('index', axis=1)
```
