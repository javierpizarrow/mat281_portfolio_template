EL conjunto de datos se denomina `company_sales_data.csv`, el cual contiene información tal como: número del mes, unidades, precio, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","company_sales_data.csv"))
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
sns.set(rc={'figure.figsize':(10,8)})

sns.lineplot(
    x='month_number',
    y='total_profit',
    #hue='Generation',# color por Generation
    data=df,
    ci = None,
)   

```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.scatterplot(
        x='month_number',
        y='total_profit',
        data=df,
        ci=None
    
)
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(
df.month_number, # eje x
df.facecream, # eje y
label='facecream', # etiquetado
color='black', # color
linewidth=1 # tamano de la curva
)

plt.plot(
df.month_number, # eje x
df.facewash, # eje y
label='facewash', # etiquetado
color='green', # color
linewidth=1 # tamano de la curva
)

plt.plot(
df.month_number, # eje x
df.toothpaste, # eje y
label='toothpaste', # etiquetado
color='blue', # color
linewidth=1 # tamano de la curva
)

plt.plot(
df.month_number, # eje x
df.bathingsoap, # eje y
label='bathingsoap', # etiquetado
color='pink', # color
linewidth=1 # tamano de la curva
)

plt.plot(
df.month_number, # eje x
df.shampoo, # eje y
label='shampoo', # etiquetado
color='brown', # color
linewidth=1 # tamano de la curva
)

plt.plot(
df.month_number, # eje x
df.moisturizer, # eje y
label='moisturizer', # etiquetado
color='Yellow', # color
linewidth=1 # tamano de la curva
)

plt.legend() # agregar etiquetado
plt.title("Ventas detalle de producto por mes") # agregar titulo
plt.xlabel('mes') # nombre eje x
plt.ylabel('cantidad de producto') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar grafico
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
fig = plt.figure(figsize=(12, 8))
plt.bar(df.month_number+0.0,df.facecream,color='blue',width=0.30,label='facecream')
plt.bar(df.month_number+0.30,df.facewash,color='red',width=0.300,label='facewash')
plt.legend()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
datos=df.drop(['month_number','total_units', 'total_profit'], axis=1)
sns.boxplot(data=datos)
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
lista1=[]
lista2=[]
lista1.append(df.facecream.sum())
lista2.append('facecream')
lista1.append(df.facewash.sum())
lista2.append('facewash')
lista1.append(df.toothpaste.sum())
lista2.append('toothpaste')
lista1.append(df.bathingsoap.sum())
lista2.append('bathingsoap')
lista1.append(df.shampoo.sum())
lista2.append('shampoo')
lista1.append(df.moisturizer.sum())
lista2.append('moisturizer')
plt.pie(lista1, labels=lista2, autopct="%0.1f %%")
plt.show()
```
